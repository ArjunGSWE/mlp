```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic graph

* plot(x, y) plots a graph with x and y axis where x and y are 1D array
* plt.show() is used to show the graph

```python
plt.plot([1, 2, 3, 4], [2, 1, 6, 0])
plt.show()
```

```python
x = [2, 4, 6, 8, 10, 12, 14]
y = [19, 2, 42, 34, 41, 0, 1]
plt.plot(x, y)
plt.show()
```

### Title
* plt.title is used to add a **title**ls to our graph

```python
plt.title("Just a GRAPH")
x = [2, 4, 6, 8, 10, 12, 14]
y = [19, 2, 42, 34, 41, 0, 1]
plt.plot(x, y)
plt.show()
```

### Labels
* to add labels to the axes of our graph

```python
x = [1, 2, 3, 4, 5, 6]
y = [2, 4, 6, 8, 10, 12]
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.plot(x, y)
plt.show()
```
